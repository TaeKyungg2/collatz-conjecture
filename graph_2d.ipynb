```python
import pandas as pd

with open('col_10k.txt', 'r', encoding='utf-8') as file:
    data=file.readlines()
data[:2]
```

```python
col_nums=list(map(lambda line : line[:-3].split(","),data))
col_nums[:2]
```

```python
col_nums=list(map(lambda line : list(map(lambda x : int(x), line)),col_nums))
col_nums[:2]
```

```python
type(col_nums[-1][1])
```

```python
from math import sin,cos,pi
def nums_to_point(num):
    x_points=[0]
    y_points=[0] 
    dir=0
    index=1
    for n in num[1:]:
        if n%2==0: 
            dir+=pi/30
            x_points.append(x_points[index-1]+sin(dir))
            y_points.append(y_points[index-1]+cos(dir))
        else:
            dir-=pi/30
            x_points.append(x_points[index-1]+sin(dir))
            y_points.append(y_points[index-1]+cos(dir))
        index+=1
    return x_points,y_points
points_1k=[] #그래프 그리는 시간 오래 걸려서 1000개만 만들기
for line in col_nums[:1000]:
    points_1k.append(nums_to_point(line))
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
plt.clf() 
plt.style.use("dark_background")
plt.figure(figsize=(10, 10))
for idx,i in enumerate(points_1k):
    color=cm.cividis(np.random.rand())
    #color=np.random.rand(4,)
    plt.plot(i[0],i[1],color=color,linewidth=0.5,alpha=0.7)
    plt.axis("equal")
name="10civ"
plt.savefig("graph/"+name+".svg")
plt.savefig("graph/"+name+".png")
plt.show()
```
